# LLM Society Quickstart (Tiny LLM Network)

A minimal, runnable example to see the package in action:
- LLM-driven scoring with a small network (n < 10)
- Sparse connectivity (low mean degree)
- 10 rounds to get a quick sense of dynamics


## 1) Setup
- Python 3.10+
- Install: `pip install -r requirements.txt`
- Provide OpenAI key: `export OPENAI_API_KEY=...` or put it in `api-key.txt` (first line)


```python
import os
os.environ['OPENAI_API_KEY'] = "<YOUR_OPENAI_API_KEY>"
```

```python
from llm_society.api import network
from llm_society.viz import set_theme

# Beautiful plots across the notebook
set_theme()
```

## 2) Create a tiny LLM network
- n=8 nodes, mean degree=2 (sparse), 10 rounds
- Default metric is `credibility` with 0–1 scores
- talk_prob controls how often pairs discuss the claim


```python
net = network(
    information="'Green roofs' substantially reduce urban heat.",
    n=8,           # small graph
    degree=2,      # sparse connectivity
    rounds=10,     # keep it short
    talk_prob=0.3, # probability a sampled pair discusses the claim
    mode="llm",    # LLM-driven metric updates
    rng=42,
)

net.simulate()
print(net.summary_report())
```

## 3) Visualize
- Final node scores on the graph
- Coverage over time
- Centrality vs final score


```python
# Final scores heatmap on the graph
net.plot(type="final")

# Coverage curve
net.plot(type="coverage")
```
